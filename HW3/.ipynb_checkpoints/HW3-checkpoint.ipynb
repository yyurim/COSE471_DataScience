### 1-3. Data  Filtering
### Filter rows with the most recent day of the column 'Date' in covid data set
- __Hint, If you are not familiar with data filtering using pandas.dataframe, Please refer to the following link below  . __
    - __https://pythonspot.com/pandas-filter/__
- __Condition 1 : Check when the 'Date' columns starts and ends using python function using  <u>covid_train4 dataset<u>__

- __Condition 3 : Filtering rows with six unique countries by "CountryName' using <u>OxCGRT_latest dataset<u>__
    
    ('CountryName' == Australia, China, Japan, New Zealand, Singapore, South Korea)

```python
#Condition 1
covid_start = covid_train.groupby('Country_Region')['Date'].min()
covid_end = covid_train.groupby('Country_Region')['Date'].max()

covid_start_df = pd.DataFrame({'Country_Region':covid_start.index,'starts':covid_start.values})
covid_end_df = pd.DataFrame({'Country_Region':covid_end.index,'ends':covid_end.values})
covid_dates = covid_start_df.merge(covid_end_df,on='Country_Region')
print("Date starts >> ",covid_start.min())
print("Date ends >> ",covid_end.max())
covid_dates
```

```python
#Condition 2
covid_dates.rename(columns = {'ends' : 'Date'}, inplace = True)
covid_dates.drop(columns='starts', inplace = True)
covid_train=pd.merge(covid_train,covid_dates)
covid_train
```

## 2. Data Visualization

### 2-1.  Plot barchart of  top10 'ConfirmedCases' countries in Merged dataset dataset in 1-5 

* example 1 
* example 2 

```python
# dataframe of top10 'ConfirmedCases' countries
        
covid.sort_values('ConfirmedCases',ascending=False, inplace=True)
confirmed10 = covid.head(10)
confirmed10
```

```python
#visualize the barchart of top 10 contries' 'ConfirmedCases' and 'Fatalities' like example 1

def compute_pos(xticks, width, i):
    # xticks is 10
    # as there are 10 groups : 10 countries with the highest death rate
    index = np.arange(xticks)
    
    # bars per group is 2
    # as there are 2 labels : ConfirmedCases, Fatalities
    n = 2
    loc = i-0.5*(n-1)
    return index + width*loc

width = 0.3

plt.figure(figsize=(17,8))
pos = compute_pos(10, width, 0)
plt.bar(pos,confirmed10['ConfirmedCases'],label='ConfirmedCases',width=width*0.95)

pos = compute_pos(10, width, 1)
plt.bar(pos,confirmed10['Fatalities'], label='Fatalities', width=width*0.95)

plt.xticks(np.arange(10),confirmed10['Country_Region'])
plt.show()
```

```python
#visualize the barchart of top 10 contries' 'ConfirmedCases' and 'Fatalities' like example 2

plt.figure(figsize=(17,8))
plt.bar(confirmed10['Country_Region'], confirmed10['ConfirmedCases'],label='ConfirmedCases')
plt.bar(confirmed10['Country_Region'], confirmed10['Fatalities'],bottom=confirmed10['ConfirmedCases'],label='Fatalities')
plt.legend(fontsize=25)
plt.show()
```

### 2-2.  Visualize your data to express 'corona infection situation globally'
#### if you are not familiar with folium package, refer to the website written at the beginning of the document.

* example 

```python
#Visualize your data to express 'corona infection situation globally' using folium package

m = folium.Map()
for i in range(len(covid)):
    confirmed = covid.loc[i,['ConfirmedCases']].values[0]
    rad = 40*(confirmed)/(confirmed10['ConfirmedCases'].values[0])
    folium.CircleMarker(location = covid.loc[i,['lat','lon']], radius = rad,color='red',fill=True,weight=4).add_to(m)
m.fit_bounds([[73,-156], [-55,177]])
m
```

### 2-3.  Visualize your data to express 'corona infection situation ' according to date
#### if you are not familiar with cufflinks package, refer to the website written at the beginning of the document.
* example 1 
* example 2 
* example 3 

```python
#load covid_for_visualize file using pandas

covid4vis = pd.read_csv('data/covid_for_visualize.csv',header=0)
covid4vis
```

```python
#visualize line graph like example 1

covid4vis = covid4vis.melt(id_vars=['Date'])
fig = px.line(covid4vis, x="Date", y="value", color="variable")
fig.update_layout(shapes=[
    dict(type= 'line',x0= 20200215,y0=-100,x1=20200215,y1=90000,line=dict(color='RED',width=1)),
    dict(type= 'line',x0= 20200325,y0=-100,x1=20200325,y1=90000,line=dict(color='RED',width=1)),
    dict(type= 'line',x0= 20200425,y0=-100,x1=20200425,y1=90000,line=dict(color='RED',width=1)),
    
])
fig.update_yaxes(title_text=None)
fig.update_xaxes(showticklabels=False, title_text=None)
```

```python
#visualize line graph like example 2

fig = px.area(covid4vis, x="Date", y="value", color="variable"
             ,facet_row="variable")
fig.for_each_annotation(lambda a: a.update(text=''))
fig.update_yaxes(showticklabels=True , matches=None, title_text=None)
fig.update_xaxes(showticklabels=False , title_text=None)
```

```python
#visualize line graph like example 3

fig = px.line(covid4vis, x="Date", y="value", color="variable",facet_col="variable",facet_col_wrap=2)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.update_yaxes(showticklabels=True , matches=None, title_text=None)
fig.update_xaxes(showticklabels=True , title_text=None)
```

***YOUR DESCRIPTION HERE***

The graph above show the connection between the number of cofirmed cases and that of fatalties of COVID19 per country. Bubbles of the graph means density of each country, and color of them mean the continent of the country.
It is obvious faltalities are proportional to confirmed cases. Africa shows relatively lower number of confirmed cases than others except Oceania, as the bubbles of Africa gathered densely between 10 and 100 confirmed cases. Most bubbles of Asia is located below the bubbles of other continent. It means Asia often has lower proportion of fatalities to confirmed cases than others. What surprise is the number of confirmed cases of each country is not actually proportional to the density of it. Rather relatively big-sized bubbles more appears below 1000 confirmed cases. And they also have lower number of fatalities that small-sized bubbles do.
