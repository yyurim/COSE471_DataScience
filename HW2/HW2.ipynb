```python
"""Parse html and return html tables of wikipedia coronavirus_pandemic data."""

#step1: look at html source!"
#step2: get all tables"
#step3: get table we care about"
    
tables = page.find_all('table',{'id':'thetable'})
table = tables[0]
table
```

## 2. Extracting data and filling it into a dictionary(Python data stucture)

#### Extract COVID-19 data for countries from table and store it in dictionary.
*Please print parsed dictionary at the end of code*

##### Shape of Dictionary
- { 'country_name': { 'cases' : 'num', 'deaths' : 'num', 'recovers' : 'num' }}
- *Dictionary data structure does not have order between Keys, so it may show up in different order(index).  You don't need to be worried of it*
- e.g. 'Uzbekistan': {'cases': '2,017', 'deaths': '9', 'recovers': '1,096'}, 'Puerto Rico': {'cases': '1,539', 'deaths': '92', 'recovers': '—'} ...

```python
# extract headings and actual data and combine data into single dict
dict_COVID19 = {}
rows = table.find_all('tr')
for i, r in enumerate(rows):
    if i>228:
        break
    if i < 2:
        continue
    country = r.find_all('th')[1].get_text().strip().split('[')[0]
    data = r.find_all('td')
    dict_COVID19[country] = {'cases':data[0].get_text().strip(), 'deaths':data[1].get_text().strip(),'recovers':data[2].get_text().strip()}
    
dict_COVID19
```

```python
#step1: create dataframe(indices: country_name, columns: cases, deaths, recovers)
#step2: replace '—' to 0

# Use the following code
df = pd.DataFrame.from_dict(dict_COVID19, orient='index')
df.loc[df['cases']=='No data','cases'] = '0'
df.loc[df['deaths']=='No data', 'deaths'] = '0'
df.loc[df['recovers']=='No data','recovers'] = '0'
df
```

```python
#step1: calculate death rates(to calculate death rates, strings in dataframe should be converted to numeric datas)
#step2: add "death_rates" column to dataframe 
#step3: sort based on death rates 

death_rates_list = []
for i, index in enumerate(df.index):
    df.loc[index]['cases'] = int(df.loc[index]['cases'].replace(',',''))
    df.loc[index]['deaths'] = int(df.loc[index]['deaths'].replace(',',''))
    df.loc[index]['recovers'] = int(df.loc[index]['recovers'].replace(',',''))
    d_rate = float(df.loc[index]['deaths']) / df.loc[index]['cases']
    death_rates_list.append(float(format(d_rate,".2f")))
df['death_rates'] = death_rates_list
df.sort_values('death_rates',ascending=False, inplace=True)
df
```

## 5-2. Plot Grouped bar chart with labels using dataframe that created from problem 5-2.

```python
import numpy as np

def compute_pos(xticks, width, i):
    # xticks is 5
    # as there are 5 groups : 5 countries with the highest death rate
    index = np.arange(xticks)
    
    # bars per group is 3
    # as there are 3 labels : cases, deaths, recovers
    n = 3
    loc = i-0.5*(n-1)
    return index + width*loc

# Show the exact number of patients above the bars
def present_height(ax, bar):
    for rect in bar:
        height = rect.get_height()
        x = rect.get_x()+rect.get_width()*0.5
        y = height*1.05
        if height == 0:
            y = 0.75
        ax.text(x, y, height, rotation=0, ha='center', va='bottom')

# Use the following code
df_highest = df.head(5)
labels = df_highest.index
cases = df_highest['cases']
deaths = df_highest['deaths']
recovers = df_highest['recovers']


plt.figure(figsize=(17,5))

width = 0.15

pos = compute_pos(5, width, 0)
case = plt.bar(pos,cases,label='cases',width=width*0.95)
present_height(plt, case)

pos = compute_pos(5, width, 1)
death= plt.bar(pos,deaths, label='deaths', width=width*0.95)
present_height(plt, death)

pos = compute_pos(5, width, 2)
recover = plt.bar(pos, recovers,label='recovers', width=width*0.95)
present_height(plt, recover)

plt.yscale('log')

plt.title('number of patients by country')
plt.ylabel('number of patients')

plt.legend(loc='best', shadow=True)

plt.xticks(np.arange(5),labels)
plt.show()
```
